```python
import matplotlib.pyplot as plt
from os import listdir
import numpy as np
plt.style.use("seaborn")
```

```python
base_path = "../data/v2_distilled_indic_en"
base_path_HQ = "../data/v2_distilled_indic_en_HQ"
hq_total, total, eps = 0, 0, 0.75

for lang_pair in listdir(base_path):
    lang = lang_pair.split('-')[1]
    with open(f"{base_path}/{lang_pair}/train.en") as f, \
         open(f"{base_path_HQ}/{lang_pair}/train.en") as f_HQ:
        data = np.array([len(s) for s in f.readlines()])
        data_HQ = np.array([len(s) for s in f_HQ.readlines()])

    u, s = np.mean(data), np.std(data)
    u_HQ, s_HQ = np.mean(data_HQ), np.std(data_HQ)

    fig, (ax1, ax2) = plt.subplots(2, sharex=True)
    plt.figure(figsize=(20, 20), dpi=120)
    fig.tight_layout(pad=5.0)
    fig.suptitle(f"Language: {lang}")
    ax1.hist(data, bins="auto")
    ax1.axvline(x=u, color = 'r')
    ax1.set_xlabel('freq', fontsize=10)
    ax1.set_ylabel('lens', fontsize=10)
    ax1.set_xlim(0, 250)
    ax1.set_title(f"sentence lens of data\nMean: {u:.3f}, Std: {s:.3f}")

    ax2.hist(data_HQ, bins="auto")
    ax2.axvline(x=u_HQ, color = 'r')
    ax2.set_xlabel('freq', fontsize=10)
    ax2.set_ylabel('lens', fontsize=10)
    ax1.set_xlim(0, 250)
    ax2.set_title(f"sentence lens of data_HQ\nMean: {u_HQ:.3f}, Std: {s_HQ:.3f}")
    plt.show()
```
